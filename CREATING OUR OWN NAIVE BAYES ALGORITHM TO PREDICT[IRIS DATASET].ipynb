```python
# FIRST WE WILL FIT THE MODEL AND THEN WE PREDICT IT
import numpy as np
import pandas as pd
```

```python
# THIS IS FOR FIT FUNCTION
def fit(X_train,Y_train):
    #IT TAKES X_TRAIN AND Y_TRAIN AS PARAMETERS
    # WE WILL CREATE DICTIONARY TO STORE ALL THE CLASS AS KEYS , FEATUES AND ITS NAMES AS VALUES
    result={}
    # GET THE CLASSESS
    allclasses=set(Y_train)
    # THIS WILL PROVIDE YOU THE CLASS
    #ITERARE THROUGH THE CLASSES
    for class_ in allclasses:
        result[class_]={}
        # AGAIN CREATING THE DICTIONARY TO STORE THE FEATURES AS KEYS
        result['total_count']=len(Y_train)
        current_class=(Y_train==class_)
        X_train_current=X_train[current_class]
        Y_train_current=Y_train[current_class]
        # YOU ALSO HAVE TO STORE THE COUNT OF EACH CLASS ,CLASS MEANS THE OUTPUT VALUES[YES/NO]
        result[class_]['count']=len(Y_train_current)
        # NOW YOU HAVE TO GET ALL THE FEATURES FROM THAT
        allfeatures=X_train.shape[1]
        for j in range(1,allfeatures+1):
            result[class_][j]={}
            # TO STORE THE UNIQUE VALUE WHCIH WILL BE PRESENT IN THAT PARTICULAR FEATURE
            #NOW YOU GET THE NAMES OF THAT FEATURES
            allnames=set(X_train[:,j-1])
            # TRAVERSE THEOUGH THE NAME
            for name in allnames:
                result[class_][j][name]=(X_train_current[:,j-1]==name).sum()
                # YOU WILL STORE ALL THE NAMES IN THE DICTIOANRY
    return result
```

```python
# IN THIS WE ARE TAKING THE DICTIONARY AND WE ARE TAKING THE X_TEST
```

```python
# NOW YOU WRITE THE FUNCTION FOR SINGLE PREDICTION /SINGLE TEST DATA PREDICTION
def single_point_prediction(dictionary,x,class_):
    output=dictionary[class_]['count']/dictionary['total_count']
    # YOU HAVE TO GET THE NUMBER OF FEATURES YOU CAN DO THAT IN TWO WAYS
    #numoffeatures=x.shape[-1]# WILL GIVE YOU THE NUMBER OF FEATURES
    numoffeatures=len(dictionary[class_].keys())-1
    # THIS WILL GIVE YOU THE FEATURES WHICH IS WITH THAT PARTICULAR CLASS
    # NOW ITERATE THROUGH THE LOOP
    for i in range(1,numoffeatures+1):
        # NOW YOU GET THE VALUES OR NAMES FROM THAT FEATURES
        xj=x[i-1]
        # THIS IS LAPLACE TRANSFORMATION TO AVOID ZERO PROBABILITIES
        counttt=dictionary[class_][i][xj]+1
        denomi=dictionary[class_]['count']+len(dictionary[class_][i].keys())
        answer=np.log(counttt)-np.log(denomi)
        # WE ARE USING LOG TO AVOID OUTPUT TO BECOME 0 IF IT HAS LOT MORE FEATURES IT HAS A TENDENCY TO BECOME ZERO
        output=output+answer
    return output
```

```python
# NOW YOU ARE WRITING A FUNCTION TO PREDICT EACH DATA FROM X_TEST
def predictions(dictionary,x):
    # YOU HAVE TO GET ALL THE CLASSES FROM THE DICTIONARY [CLASSES MEANS  EX:YES/NO]
    classes=list(dictionary.keys())# YOU WILL GET ALL THE CLASSES
    bestprob=-10000
    bestclass=-1
    # WE WILL INITILIAZE BOTH BEST PROB AND BEST CLASS IN NEGATIVE VALUE BECAUSE PROB AND BEST CLASS CANNOT BE IN NEGATIVE
    firstrun=True
    # NOW YOU ITERATE THROUGH ALL THE CLASSESS
    for class_ in classes:
        # THIS WILL GIVE YOU 'total_count' as one of the class IF IT ENCOUNTERS JUST CONTINUE IT
        if class_=='total_count':
            continue
        bestclprob=single_point_prediction(dictionary,x,class_)
        if firstrun or bestclprob>bestprob:
            bestprob=bestclprob
            bestclass=class_
        firstrun=False
        # YOU HAVE DONE THE FIRST RUN SO TURN IT TO NEGATIVE
    # YOU RETURN WHATEVER BEST CLASS IT WILL GIVE YOU
    return bestclass

            
            
```

```python
def predict(dictionary,X_test):
    # WE ARE STORING ALL THE BEST CLASS Y_PRED IN THE LISTS
    y_pred=[]
    # WE ARE TRAVERSING THROUGH EACH DATA OF THE DATASET
    for x in X_test:
        #print(x)
        # WE ARE CALLING ANOTHER PREDICT FUCNTION TO PREDICT EACH DATA
        best_Class=predictions(dictionary,x)
        y_pred.append(best_Class)
    # FINALLY WE ARE RETURNING Y_PRED WHICH HAS ALL THE PREDICTIONS
    return y_pred
```

```python
def powerr(column):
    seco=column.mean()
    fi=0.5*seco
    ti=1.5*seco
    for i in range(0,len(column)):
        if column[i]<fi:
            column[i]=0
        elif column[i]<seco:
            column[i]=1
        elif column[i]<ti:
            column[i]=2
        else:
            column[i]=3
    return column
```

```python
from sklearn import datasets
iris=datasets.load_iris()
X=iris.data
Y=iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
print(X_train)
print(Y_train)
print(X_test)
print(Y_test)
```

```python
dictionary=fit(X_train,Y_train)
```

```python
Y_pred=predict(dictionary,X_test)
```

```python
print(Y_pred)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(Y_test,Y_pred))
```

```python
print(confusion_matrix(Y_test,Y_pred))
```
